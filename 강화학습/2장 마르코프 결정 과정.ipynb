- 2 

```python
# 2  L1  
V = 1

for i in range(1, 100):
    if i % 2 == 1:
        V += 0
    else :
        V += 1 * (0.9 ** i)
print("L1  , 2  : ", V)
```

```python
# 2  L2  
V = 0

for i in range(1, 100):
    if i % 2 == 1:
        V += 1 * (0.9 ** i)
    else :
        V += 0
print("L2  , 2  : ", V)
```
