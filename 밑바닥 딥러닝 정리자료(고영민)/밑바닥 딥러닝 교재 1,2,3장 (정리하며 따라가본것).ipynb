## 2.3 퍼셉트론 구현하기

### 2.3.1. AND 게이트


```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7 # 사람이 정해줘야 함.
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta :
        return 0
    elif tmp > theta :
        return 1
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

### 2.3.3. 가중치와 편향 구현하기

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else :
        return 1
```

```python
print(AND(x1=0, x2=0))
print(AND(x1=1, x2=0))
print(AND(x1=0, x2=1))
print(AND(x1=1, x2=1))
```

- NAND 게이트와 OR 게이트 구현

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0 :
        return 0
    else : 
        return 1
```

```python
print(NAND(x1=0, x2=0))
print(NAND(x1=1, x2=0))
print(NAND(x1=0, x2=1))
print(NAND(x1=1, x2=1))
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else :
        return 1
```

```python
print(OR(x1=0, x2=0))
print(OR(x1=1, x2=0))
print(OR(x1=0, x2=1))
print(OR(x1=1, x2=1))
```

## 2.4 퍼셉트론 한계

### 2.5.2. XOR 게이트 구현

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
print(XOR(x1=0, x2=0))
print(XOR(x1=1, x2=0))
print(XOR(x1=0, x2=1))
print(XOR(x1=1, x2=1))
```

- ReLU(Rectified Linear Unit) 렐루 함수 : 정류된 선형 함수   
- 입력이 0보다 크면 그대로 출력, 아니면 모두 0을 출력

```python
def relu(x):
    return np.maximum(0, x)
```

## 3.4. 3층 신경망 구현하기

- 행렬과 벡터가 선형대수학에서 정의된 연산과 다르게 넘파이에서는 몇 가지 테크닉(브로드캐스트 등)을 사용하여 구현함. 이 점에 유의 해야함.

#### Step 1. 1번 째 층 선형변환

```python
X = np.array([1.0, 0.5])
X = np.reshape(X, (1, -1))
W1 = np.array([[0.1, 0.3, 0.5],
              [0.2, 0.4, 0.6]])
B1 = np.array([0.1, 0.2, 0.3])
B1 = np.reshape(B1, (1, -1))

print(np.shape(X))
print(np.shape(W1))
print(np.shape(B1))
```

```python
print(np.dot(X,W1))
```

```python
A1 = np.dot(X, W1) + B1
print(A1)
print(np.shape(A1))
```

#### Step 2. 1번째 층 선형변환 후 비선형변환(활성함수)

```python
Z1 = sigmoid(A1)
print(Z1)
```

- 활성함수는 원소곱 연산임

#### Step 3. 2번째 층의 선형변환

```python
W2 = np.array([[0.1, 0.4],
              [0.2, 0.5],
              [0.3, 0.6]])
B2 = np.array([0.1, 0.2])
B2 = np.reshape(B2, (1, -1))

print(np.shape(Z1))
print(np.shape(W2))
print(np.shape(B2))
```

#### Step 4. 2번째 층의 선형변환 후 비선형변환

```python
A2 = np.dot(Z1, W2) + B2
print(A2)
print(np.shape(A2))

Z2 = sigmoid(A2)
print(Z2)
print(np.shape(Z2))
```

```python
def forward(network, x):
    W1, W2 ,W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = identity_function(a3)
    
    return y
```

## 3.5. 출력층 설계하기

- 소프트맥스(softmax) 함수는 분류모델에서 모델의 예측값이 얼마나의 확률로 그 클래스에 속하는지를 표현하기 위해 사용하는 함수임

### 3.5.1. 소프트맥스 함수 구현

```python
a = np.array([0.3, 2.9, 4.0])
exp_a = np.exp(a)
print(exp_a)
```

```python
sum_exp_a = np.sum(exp_a)
print(sum_exp_a)
```

```python
y = exp_a / sum_exp_a
print(y)
```

- 각 확률을 다 합치면 1 을 산출함.

```python
print(np.sum(y))
```

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([-3, 1, 4])
print(softmax(a))
print(np.round(softmax(a), 3))
```

### 3.5.2. 소프트맥스 함수 구현시 주의점

- 소프트맥스 함수에 입력으로 받는 값을 지수함수 exp 을 통해 내뱉을 때 기하급수적으로 값이 커지는 경향으로 컴퓨터의 메모리가 받쳐주지 못해 nan을 뱉는 오버플로우 문제가 발생함

```python
a = np.array([1010, 1000, 990])
print(softmax(a))
```

- 한 가지 해결책으로 배열 내에 가장 큰 값을 각 배열 원소에 빼줌으로써 정규화시켜줄 수 있음

```python
c = np.max(a)
print(c)
```

```python
print(a-c)
```

```python
print(softmax(a-c))
print(np.round(softmax(a-c), 3))
```

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([1010, 1000, 990])
print(np.round(softmax(a), 3))
```
